### Performing Univariate Analysis

 Examining one variable at a time to gain insights into the data and identify patterns or anomalies that may require further investigation.

Considering the 'Age' Attribute 

```python
#Calculating the distribution of values
age_frequency = data['Age'].value_counts()

age_percentage = (age_frequency / len(data)) * 100
print("Percentage of each age category:")
print(age_percentage)
```

```python
plt.figure(figsize=(5, 3))
sns.histplot(data['Age'], kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

The value of the 'Age' attriute can be discretized to gain insight on the shopping behaviour of Age Groups, rather than a particular age.

```python
#Defining the bins for age groups
bins = [18, 30, 40, 50, 60, 70, 120]

#Defining the labels for the age groups
labels = ['18-29', '30-39', '40-49', '50-59', '60-69', '70+']

#Discretizing the 'Age' variable into age groups (creating a seperate column 'Age Group')
data['Age Group'] = pd.cut(data['Age'], bins=bins, labels=labels, right=False)
```

```python
#Checking the distribution post discretization
age_group_frequency = data['Age Group'].value_counts()

age_group_percentage = (data['Age Group'].value_counts() / len(data)) * 100
print("Percentage of each age category:")
print(age_group_percentage)
```

```python
#Visualizing the result
plt.figure(figsize=(5, 3))
sns.countplot(x='Age Group', data=data)
plt.title('Frequency of Age Categories')
plt.xlabel('Age Group')
plt.ylabel('Frequency')
plt.xticks(rotation=45)
plt.show()
```

### Performing Bivariate Analysis

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(data['Gender'], data['Shipping Type'])

#Performing chi-square test
chi2_stat, p_val, dof, expected_values = chi2_contingency(contingency_table)

#Results
print("Chi-square statistic:", chi2_stat)
print("P-value:", p_val)
print("Degrees of freedom:", dof)
print("Expected frequencies:")
print(expected_values)
```

```python
import statsmodels.graphics.mosaicplot as mosaic

#Creating the mosaic plot
plt.figure(figsize=(8, 6))
mosaic.mosaic(data, ['Gender', 'Shipping Type'], title='Mosaic Plot: Gender vs. Shipping Type')
plt.show()
```

```python
from scipy.stats import f_oneway

#Extracting purchase amounts for different seasons
purchase_winter = data[data['Season'] == 'Winter']['Purchase Amount (USD)']
purchase_spring = data[data['Season'] == 'Spring']['Purchase Amount (USD)']
purchase_summer = data[data['Season'] == 'Summer']['Purchase Amount (USD)']
purchase_fall = data[data['Season'] == 'Fall']['Purchase Amount (USD)']

#Performing one-way ANOVA
f_statistic, p_value = f_oneway(purchase_winter, purchase_spring, purchase_summer, purchase_fall)

#Printing the results
print("One-way ANOVA Results:")
print("F-statistic:", f_statistic)
print("p-value:", p_value)

#Interpreting the results
alpha = 0.05
if p_value < alpha:
    print("There is a significant difference in average purchase amounts across different seasons.")
else:
    print("There is no significant difference in average purchase amounts across different seasons.")
```
